## Web scraping 

- **Doctors of Manipal Hospital**

```
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```
base_url = 'https://www.manipalhospitals.com/saltlake/doctors-list/'

header = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36",
}
```

```
doc_name = []
Consultant = []
Qualification = []


for page_num in range(1,10):
    url = f"{base_url}{page_num}/"
    response = requests.get(url, headers = header).text

    hersoup = BeautifulSoup(response, 'lxml')

    company = hersoup.find_all('div', 'doc-profile-appoint doc-listing-appointment')


    for i in company:
        doc_name.append(i.find('h3', 'doctor_name_fetch').text.strip())
        Consultant.append(i.find('p', 'desgn').text.strip())
        Qualification.append(i.find("tbody", class_="table-qualification").find("p").text.strip())


table = pd.DataFrame({'doc_name' : doc_name, 'Consultant' : Consultant, 'Qualification' : Qualification})
```

- **Doctors of Apollo Hospital**

```
url = 'https://www.apollohospitals.com/book-doctor-appointment/?page='

headerx =  {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36",
}
```

```
Doctor_Name = []
Department = []
Experince = []
Availablity = []
Timings = []


for page_no in range(1,50):
    urlx = f"{url}{page_no}"

    responsex = requests.get(urlx, headers= headerx).text

    mysoup = BeautifulSoup(responsex, 'lxml')

    network = mysoup.find_all('div', 'card-body p-lg-4 p-3 text-black')

    for i in network:
        Doctor_Name.append(i.find('h2', 'mb-1 textline1 ft-wight-1').text.strip())
        Department.append(i.find('p', 'mb-0 me-2 textline2 akscolr4 fs013 heightline-1').text.strip())
        Experince.append(i.find('p', 'mb-0 me-2 dp-col2 textline1 fs013').text.strip())
        Availablity.append(i.find('span', 'doc-weekdays').text.strip())
        Timings.append(i.find('span', 'doc-timings').text.strip())


table2 = pd.DataFrame({'Doctor_Name' : Doctor_Name, 
                       'Department' : Department,
                       'Experince' : Experince,
                       'Availablity' : Availablity,
                       'Timings' : Timings}) 
```
